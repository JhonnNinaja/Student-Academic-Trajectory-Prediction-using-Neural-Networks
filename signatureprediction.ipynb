```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import sklearn as sk
import sklearn.neural_network

from sklearn.neighbors import KNeighborsRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.datasets import make_multilabel_classification

from sklearn.multioutput import MultiOutputClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv(r"/content/DataSet.csv")
df.head()
```

```python
data = np.array(df.values)

s = 50 #semestres
m = 49 #materias
In_train = np.zeros((s, m)) #(semestres, materias)
Out_train = np.zeros((s, m)) #(semestres, materias)

ind_count = lambda _ix: np.count_nonzero(In_train[_ix,2:] == 1)

for ix, i in enumerate(data[:]):

  In_train[ix] = data[ix,2:]

  if ix - 1 < 0:
    Out_train[ix] = data[ix,2:]
  else:
    if ix % 10 == 0:
      Out_train[ix] = data[ix,2:]
    else:
      Out_train[ix] = data[ix,2:] - data[ix-1,2:]


for ix, i in enumerate(Out_train[:-1]):
  Out_train[ix] = Out_train[ix + 1]
  if ix == m-1:
    Out_train[m] = np.zeros(m)

```

```python
#nn = [49, 10, 12, 10, 49]
#learning_rate = 2

neural_net = MultiOutputClassifier(KNeighborsClassifier())
#model = sk.neural_network.MLPRegressor((10,49), random_state = 0, learning_rate_init = 0.001, activation = "logistic",max_iter = 2000000)

knn = KNeighborsRegressor()
regr = MultiOutputRegressor(knn)
```
